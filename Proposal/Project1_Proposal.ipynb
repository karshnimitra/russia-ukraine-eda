## Data:

Our data is currently consists of the following Kaggle datasets:
- https://www.kaggle.com/datasets/5481b37f75c2bc8587e527d6e4d6fd7a81ac46e329e1946c8f85210cbe662eb5, 
- https://www.kaggle.com/datasets/piterfm/2022-ukraine-russian-war

The data describes equipment losses, death toll and military wounded & prisoners of war of Russians in the 2022 Ukraine Russia War. The data has 354 observations, one for each day since the war started. This dataset is updated weekly. These datasets seem to consist primarily  of cleaned data. Later, as the project progresses, we may need to utilize additional datasets in the project.

The personnel losses dataset has 354 rows with features like date, day and number of personnel lost and number of personnel captured.

The equipment losses dataset has about 354 rows with 18 columns. The features consist of date, day, number of equipment losses. The equipments have various categories such as drones, tanks, naval ships, helicopters etc.

There are a few projects on Kaggle that utilize these datasets, but their focus is mainly on exploratory data analysis and data visualization. There seems to be no overlap between these projects and the questions we're exploring.
